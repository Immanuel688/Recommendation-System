# Here we have 1682 unique movies, which means same movies are being rated by different users, inorder to consolidate the multiple ratings given for a single movie we find the average rating 

```python
avg_ratings=ratings.groupby('movie_id')['rating'].mean().reset_index()
avg_ratings
```

```python
avg_ratings.rename(columns={'rating':'avg_ratings','movie_id':'movie id'},inplace=True)
```

```python
avg_ratings
```

```python
top_rated=avg_ratings[avg_ratings['avg_ratings']>4]
top_rated
```

```python
def popularity_based(ratings,movie_content,threshold):
    avg_ratings=ratings.groupby('movie_id')['rating'].mean().reset_index()
    avg_ratings.rename(columns={'rating':'avg_ratings','movie_id':'movie id'},inplace=True)
    #avg_ratings
    top_rated=avg_ratings[avg_ratings['avg_ratings']>threshold]
    #top_rated
    # Merging both movie and top rated dataframe
    merge= high_rated_movies = pd.merge(top_rated, movie_content,on='movie id')
    list=[col for col in movie_content.columns if col not in ['movie id', 'movie title']]
    #list=['Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
    for columns in list:
        content=merge[merge[columns]==1].sort_values(by='avg_ratings',ascending=False).head(5)['movie title'].values
        content_table=pd.DataFrame(content,columns=['movie title'])
        print('\n',"TOP 5",columns,"MOVIES:")
        print('\n',content_table)

 
```

```python
popularity_based(ratings,movie_content,3.5)
```
