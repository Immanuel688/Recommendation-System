```python
datama=ratings.pivot_table(index='user_id',columns='movie_id',values='rating')
```

Find Cosine Similatity for user and Item

```python
from sklearn.metrics.pairwise import pairwise_distances 
user_similarity = pairwise_distances(data_matrix, metric='cosine')
item_similarity = pairwise_distances(data_matrix.T, metric='cosine')
```

```python
pd.DataFrame(item_similarity)
```

Using formula for user and item we are calcuating the score value

```python
def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        #We use np.newaxis so that mean_user_rating has same format as ratings
        ratings_diff = (ratings - mean_user_rating.values[:, np.newaxis])
        pred = mean_user_rating.values[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

2. Code Components
Let’s analyze each part of the code:

a. mean_user_rating.values[:, np.newaxis]
mean_user_rating is a 1D array of shape (num_users,) containing the mean rating for each user.

[:, np.newaxis] reshapes it into a 2D array of shape (num_users, 1) for broadcasting.

b. similarity.dot(ratings_diff)
similarity is the user-user similarity matrix of shape (num_users, num_users).

ratings_diff is the matrix of rating deviations (actual ratings minus mean ratings) of shape (num_users, num_items).

The dot product similarity.dot(ratings_diff) results in a matrix of shape (num_users, num_items).

c. np.array([np.abs(similarity).sum(axis=1)])
np.abs(similarity).sum(axis=1) computes the sum of absolute similarities for each user, resulting in a 1D array of shape (num_users,).

Wrapping it in np.array([...]) converts it into a 2D array of shape (1, num_users).

d. .T (Transpose)
The transpose operation converts the 2D array from shape (1, num_users) to shape (num_users, 1).

3. Why Transpose is Needed
The division operation in the formula:

requires the shapes of the numerator and denominator to align for broadcasting.

similarity.dot(ratings_diff) has shape (num_users, num_items).

The denominator (sum of absolute similarities) needs to have shape (num_users, 1) to divide each row of the numerator by the corresponding sum of similarities.

Without the transpose:

The denominator would have shape (1, num_users).

Broadcasting would attempt to divide each column of the numerator by the denominator, which is incorrect.

With the transpose:

The denominator has shape (num_users, 1).

Broadcasting divides each row of the numerator by the corresponding sum of similarities, which is the intended behavior.
